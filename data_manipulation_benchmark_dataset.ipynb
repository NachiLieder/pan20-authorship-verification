```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


import random
p = 0.05  # 1% of the lines

df = pd.read_csv(
         '../../ml_dataset/blogtext.csv',
         skiprows=lambda i: i>0 and random.random() > p
)
df.shape
```

```python
df = df[df['text'].str.len()>200]
df.shape
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction import text
from nltk.stem import PorterStemmer

ps=PorterStemmer()

stop = stopwords.words('english')


df['text'] = df['text'].str.lower()
df['text'] = df['text'].str.replace('[^\w\s]','')

df['text'] = df['text'].apply(lambda x: ' '.join([ps.stem(word) for word in x.split() if word not in (stop)]))
```

```python
ix = df.groupby('id').size().sort_values(ascending=False)[:10].index
df = df[df['id'].isin(ix)]
df.shape
```

```python
full_df = pd.DataFrame()
for ix,row in df.iterrows():
    print(ix)
    for ix_2,row_2 in df.copy().iterrows():
        new_row = [row['text'],row_2['text'], row['id'], row_2['id']]
        full_df = full_df.append([new_row])
```

```python
full_df.shape
full_df.columns = ['text_1','text_2','id_1','id_2']
frac_df = full_df.sample(frac = 0.01)
frac_df.reset_index(drop=True, inplace=True)
frac_df['same'] = frac_df['id_1']==frac_df['id_2']

frac_df['text_id_1'] = frac_df.index
frac_df['text_id_1'] = (frac_df['text_id_1']).astype(str) + '_1'
frac_df['text_id_2'] = frac_df.index
frac_df['text_id_2'] = (frac_df['text_id_2']).astype(str) + '_2'
frac_df[['id_1','id_2','isSame','text_id_1','text_id_2']].to_csv('id_pairs_for_calc_distances_benchmark.csv',index=False)
frac_df

df_1 = frac_df[['text_1','id_1','text_id_1']]
df_2 = frac_df[['text_2','id_2','text_id_2']]
df_1.columns= ['text','id','text_id']
df_2.columns= ['text','id','text_id']
df_text_and_id = df_1.append(df_2)
df_text_and_id.shape
```

```python
full_df
frac_df
```

```python
# create frequency matrix

line_counter=1
frequency = pd.DataFrame()
# dvr_raw = list(df['text_1']) + list(df['text_2']) 
for ix, row in df_text_and_id.iterrows():
    
    if line_counter%500 == 0:
        print(line_counter)
    single_df = pd.Series(row['text'].split()).value_counts().to_frame(name='frequency_in_category').reset_index().rename(columns={'index': 'element'})
    single_df['category'] = row['text_id']
    frequency = frequency.append(single_df)
    line_counter += 1
```

```python
frequency.to_csv('frequency_nachi_benchmark.csv', index=False)
```
